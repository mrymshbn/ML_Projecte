# زمان تردد از دوربین آ به ب

```python
import pandas as pd

# خواندن داده‌ها
df = pd.read_csv('Data_p1.csv')

# مرتب‌سازی بر اساس شناسه دوربین از اول به آخر
df_sorted = df.sort_values(by='DEVICEID')

# ترتیب دوربین‌ها بر اساس شناسه DEVICEID
sorted_cameras = sorted(set(df_sorted['DEVICEID']))

# محاسبه زمان تردد بین دوربین‌ها
for i in range(len(sorted_cameras) - 1):
    camera1 = sorted_cameras[i]
    camera2 = sorted_cameras[i + 1]
    
    time_diff = df_sorted[df_sorted['DEVICEID'] == camera2]['PASSDATETIME'].values[0] - df_sorted[df_sorted['DEVICEID'] == camera1]['PASSDATETIME'].values[0]
    
    print(f"زمان تردد از دوربین {camera1} به دوربین {camera2}: {time_diff} واحد زمان")
```

## زمان تردد بین دوربین ها

```python
import pandas as pd

# خواندن داده‌ها
df = pd.read_csv('Data_p1.csv')

# مرتب‌سازی بر اساس شناسه دوربین از اول به آخر
df_sorted = df.sort_values(by='DEVICEID')

# ترتیب دوربین‌ها بر اساس شناسه DEVICEID
sorted_cameras = sorted(set(df_sorted['DEVICEID']))

# محاسبه زمان تردد بین دوربین‌ها
results = []
for i in range(len(sorted_cameras) - 1):
    camera1 = sorted_cameras[i]
    camera2 = sorted_cameras[i + 1]
    
    time_diff = df_sorted[df_sorted['DEVICEID'] == camera2]['PASSDATETIME'].values[0] - df_sorted[df_sorted['DEVICEID'] == camera1]['PASSDATETIME'].values[0]
    
    results.append({'From': camera1, 'To': camera2, 'Time': time_diff})

# تبدیل نتایج به دیتافریم
df_results = pd.DataFrame(results)

# تنظیم واحدها برای ستون زمان
df_results['Time'] = pd.to_timedelta(df_results['Time'], unit='s')

# تغییر نام ستون‌ها
df_results = df_results.rename(columns={'From': 'از دوربین', 'To': 'به دوربین', 'Time': 'زمان تردد'})

# تغییر نام سطرها
df_results = df_results.rename(index=lambda x: x + 1)

# نمایش جدول
print(df_results)
```

## نمایش اطلاعات بالا به صورت نمودار

```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن داده‌ها
df = pd.read_csv('Data_p1.csv')

# محاسبه تعداد تردد بر اساس دوربین
camera_counts = df['DEVICEID'].value_counts()

# مرتب‌سازی بر اساس تعداد تردد
sorted_cameras = camera_counts.index.sort_values()

# تنظیم اندازه نمودار
plt.figure(figsize=(18,13))

# تنظیم رنگ‌ها
colors = ['blue', 'green', 'orange', 'red', 'purple']

# رسم نمودار میله‌ای   
plt.bar(sorted_cameras, camera_counts[sorted_cameras], color=colors)

# تنظیم نام محورها  
plt.xlabel('Camera ID')
plt.ylabel('Traffic Count')

# تنظیم عنوان نمودار
plt.title('Traffic Count for Each Camera')

# نمایش نمودار
plt.show()
```
