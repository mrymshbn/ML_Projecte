## کاوش داده ها

```python
import matplotlib.pyplot as plt
import seaborn as sns
def make_hist(PASSDATETIME=1):
    data = df[df['PASSDATETIME'] == PASSDATETIME]
    f, ax = plt.subplots(figsize=(17, 5))
    ax = sns.histplot(data['vehicleId'], kde=True, stat='probability')
    ax.set_title(f'Plot show the distribution of data in PASSDATETIME {PASSDATETIME}')
    ax.grid(True, ls='-.', alpha=0.75)
    plt.show()
```

```python
make_hist(1)
```

```python
make_hist(2)
```

```python
make_hist(3)
```

```python
make_hist(4)
```
