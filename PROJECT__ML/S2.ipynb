# توصیفی از این که یک دوربین چقدر کار کرده؟ چقدر درست کار کرده؟

برای توصیف یک دوربین و میزان عملکرد آن، می‌توانید از آماره‌های مختلفی مانند تعداد تکرار حضور دوربین، مجموع زمان عبوری که دوربین رصد کرده است، میانگین زمان عبوری که دوربین رصد کرده است و سایر آماره‌های مرتبط استفاده کنید. با استفاده از کتابخانه pandas و متدهایی که در اختیار دارید، می‌توانید این آماره‌ها را برای هر دوربین محاسبه کنید.

# در زیر با چندین روش  کد قرار داده‌ام که آماره‌های مرتبط با عملکرد دوربین‌ها را محاسبه می‌کند

```python
import pandas as pd

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# محاسبه زمان نسبی از صفر
df['RELATIVE_TIME'] = df['PASSDATETIME'] - df['PASSDATETIME'].min()

# گروه‌بندی بر اساس شناسه دوربین
grouped = df.groupby('DEVICEID')

# نمایش توصیفی از هر دوربین
for group, data in grouped:
    device_id = group
    print(f"Device ID: {device_id}")
    
    # تعداد تکرار حضور دوربین
    count = data.shape[0]
    print(f"Count: {count}")
    
    # مجموع زمان عبوری که دوربین رصد کرده است
    total_pass_time = data['RELATIVE_TIME'].sum()
    print(f"Total Pass Time: {total_pass_time}")
    
    # میانگین زمان عبوری که دوربین رصد کرده است
    mean_pass_time = data['RELATIVE_TIME'].mean()
    print(f"Mean Pass Time: {mean_pass_time}")

    print()
```

## توصیف به صورت جدول و واضح

```python
import pandas as pd

# خواندن داده‌ها
df = pd.read_csv('Data_p1.csv')

# گروه‌بندی بر اساس شناسه دوربین
grouped = df.groupby('DEVICEID')

# محاسبه آماره‌های مربوط به دوربین‌ها
camera_stats = grouped['PASSDATETIME'].describe()

# توصیف جدولی از تعداد کارکرد هر دوربین
camera_usage_table = camera_stats[['count', 'mean', 'min', 'max']].rename(columns={
    'count': 'تعداد کارکرد',
    'mean': 'میانگین کارکرد',
    'min': 'کمترین کارکرد',
    'max': 'بیشترین کارکرد'
})

print(camera_usage_table)
```

## توصیف به صورت نمودار  

```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv').head(1000)

# گروه‌بندی بر اساس شناسه دوربین
grouped = df.groupby('DEVICEID')

# محاسبه آماره‌های مربوط به دوربین‌ها
camera_stats = grouped['PASSDATETIME'].describe()

# ترسیم نمودار جعبه ای
plt.figure(figsize=(21,7))
plt.boxplot(camera_stats['count'], vert=False)
plt.xlabel('Count')
plt.title('Camera Presence Statistics')
plt.show()

```

##  توصیف به صورت نمودار میله ای  

```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv').head(10000)

# گروه‌بندی بر اساس شناسه دوربین
grouped = df.groupby('DEVICEID')

# محاسبه تعداد تکرار حضور دوربین‌ها
camera_counts = grouped['PASSDATETIME'].count()

# ترسیم نمودار میله‌ای
plt.figure(figsize=(17,9))
plt.bar(camera_counts.index, camera_counts.values)
plt.xlabel('Camera ID')
plt.ylabel('Count')
plt.title('Camera Presence Counts')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# انتخاب 50 نمونه تصادفی
df_subset = df.sample(n=10500, random_state=12)

# گروه‌بندی بر اساس شناسه دوربین
grouped = df_subset.groupby('DEVICEID')

# محاسبه تعداد تکرار حضور دوربین‌ها
camera_counts = grouped['PASSDATETIME'].count()

# ترسیم نمودار میله‌ای افقی
plt.figure(figsize=(18, 9))
colors = plt.cm.viridis(np.linspace(0, 1, len(camera_counts)))
plt.barh(camera_counts.index, camera_counts.values, color=colors)
plt.xlabel('Count')
plt.ylabel('Camera ID')
plt.title('Camera Presence Counts')
plt.show()
```

## توصیفی از کارکرد صحیح هر دوربین

```python
import pandas as pd

# خواندن داده‌ها
df = pd.read_csv('Data_p1.csv')

# گروه‌بندی بر اساس شناسه دوربین
grouped = df.groupby('DEVICEID')

# محاسبه آماره‌های مربوط به زمان کارکرد صحیح هر دوربین
camera_stats = grouped['PASSDATETIME'].agg(['mean', 'min', 'max', 'count'])

print(camera_stats)
```
