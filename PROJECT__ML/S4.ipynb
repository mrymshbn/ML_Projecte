# پیدا کردن ترافیک

```python
import pandas as pd

# خواندن داده‌ها
df = pd.read_csv('Data_p1.csv')

# گروه‌بندی بر اساس دوربین‌ها و محاسبه ترافیک
traffic_data = df.groupby(['DEVICEID']).size().reset_index(name='TrafficCount')

# محاسبه مقادیر حداقل، حداکثر و میانگین
min_traffic = traffic_data['TrafficCount'].min()
max_traffic = traffic_data['TrafficCount'].max()
mean_traffic = traffic_data['TrafficCount'].mean()

# نمایش مقادیر
print(f"Minimum Traffic: {min_traffic}")
print(f"Maximum Traffic: {max_traffic}")
print(f"Mean Traffic: {mean_traffic}")
```

```python
# گروه‌بندی بر اساس دوربین‌ها و محاسبه ترافیک
traffic_data = df.groupby(['DEVICEID']).size().reset_index(name='TrafficCount')

# نمایش ترافیک بین دوربین‌ها
for index, row in traffic_data.iterrows():
    camera_id = row['DEVICEID']
    traffic_count = row['TrafficCount']
    print(f"Camera ID: {camera_id}, Traffic Count: {traffic_count}")
```

## با محاسبه میانگین،  زمان تردد برای هر دوربین، میزان تراکم ترافیک بررسی می شود 

```python
# گروه‌بندی بر اساس دوربین‌ها و محاسبه میانگین زمان تردد
camera_avg_time = df.groupby('DEVICEID')['PASSDATETIME'].mean()

# نمایش نتایج
print(camera_avg_time)
```

## نمایش محور ترافیک

```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# گروه‌بندی بر اساس دوربین‌ها و محاسبه میانگین زمان تردد
camera_avg_time = df.groupby('DEVICEID')['PASSDATETIME'].mean()

# تنظیم اندازه نمودار
plt.figure(figsize=(12, 8))

# ترتیب نمایش دوربین‌ها بر اساس میانگین زمان تردد
sorted_cameras = camera_avg_time.sort_values().index

# رسم نمودار میانگین زمان تردد برای هر دوربین
plt.bar(sorted_cameras, camera_avg_time[sorted_cameras])

# تنظیم نام محورها
plt.xlabel('Camera ID', fontweight='bold')
plt.ylabel('Average Travel Time')

# تنظیم عنوان نمودار
plt.title('Average Travel Time for Each Camera')

# چرخش نام دوربین‌ها بر روی محور x
plt.xticks(rotation=45)

# نمایش نمودار
plt.show()
```

#  مشخص کردن محور ترافیک  ، با محاسبه میانگین زمان تردد دوربین ها 

```python

# گروه‌بندی بر اساس دوربین‌ها و محاسبه میانگین زمان تردد
camera_avg_time = df.groupby('DEVICEID')['PASSDATETIME'].mean()

# ساخت جدول با اطلاعات
table = pd.DataFrame({'Camera ID': camera_avg_time.index, 'Average Travel Time': camera_avg_time.values})

#   محور ترافیک
table = table.rename_axis('Traffic Axis', axis='columns')

# نمایش جدول
print(table)
```
