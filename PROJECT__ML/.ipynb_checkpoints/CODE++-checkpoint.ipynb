```python
import pandas as pd

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# محاسبه زمان نسبی از صفر
df['RELATIVE_TIME'] = df['PASSDATETIME'] - df['PASSDATETIME'].min()

# مرتب‌سازی بر اساس زمان نسبی به صورت صعودی
df_sorted = df.sort_values('RELATIVE_TIME', ascending=True)

# انتخاب 100 نمونه از داده‌ها
df_subset = df_sorted.head(400)

# گروه‌بندی بر اساس شناسه ماشین (vehicleId)
grouped = df_subset.groupby('vehicleId')

# حرکت روی حسگرها براساس ماشین‌ها
for group, data in grouped:
    vehicle_id = group
    print(f"Vehicle ID: {vehicle_id}")

    # شمارش تکرارهای ظهور هر دوربین
    camera_counts = data['DEVICEID'].value_counts()

    # نمایش وزن براساس تکرارهای ظهور دوربین‌ها
    for camera_id, count in camera_counts.items():
        print(f"Camera ID: {camera_id}, Weight: {count}")

    print()
```

```python
import pandas as pd

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# محاسبه زمان نسبی از صفر
df['RELATIVE_TIME'] = df['PASSDATETIME'] - df['PASSDATETIME'].min()

# مرتب‌سازی بر اساس زمان نسبی به صورت صعودی
df_sorted = df.sort_values('RELATIVE_TIME', ascending=True)

# انتخاب 100 نمونه از داده‌ها
df_subset = df_sorted.head(100)

# گروه‌بندی بر اساس شناسه ماشین (vehicleId)
grouped = df_subset.groupby('vehicleId')

# حرکت روی حسگرها براساس ماشین‌ها
for group, data in grouped:
    vehicle_id = group
    print(f"Vehicle ID: {vehicle_id}")

    # تعیین دوربین‌ها
    cameras = data['DEVICEID'].unique()

    # حرکت از دوربین اول تا آخر
    for i in range(len(cameras)-1):
        camera_id1 = cameras[i]
        camera_id2 = cameras[i+1]
        print(f"Movement from Camera {camera_id1} to Camera {camera_id2}")

    print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# محاسبه زمان نسبی از صفر
df['RELATIVE_TIME'] = df['PASSDATETIME'] - df['PASSDATETIME'].min()

# انتخاب 200 نمونه تصادفی
df_subset = df.sample(n=500, random_state=1234)

# مرتب‌سازی بر اساس زمان نسبی به صورت صعودی
df_sorted = df_subset.sort_values('RELATIVE_TIME', ascending=True)

# گروه‌بندی بر اساس شناسه ماشین (vehicleId)
grouped = df_sorted.groupby('vehicleId')

# لیست وزن‌ها
weights = []

# حرکت روی حسگرها براساس ماشین‌ها
for group, data in grouped:
    # تعیین دوربین‌ها
    cameras = data['DEVICEID'].unique()

    # محاسبه وزن بین دوربین‌ها
    weight = len(cameras) - 1
    weights.append(weight)

# نمایش گرافیکی از وزن‌ها
plt.plot(range(1, len(weights)+1), weights)
plt.xlabel('Vehicle ID')
plt.ylabel('Weight')
plt.title('Camera Sequence Weight')
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# محاسبه زمان نسبی از صفر
df['RELATIVE_TIME'] = df['PASSDATETIME'] - df['PASSDATETIME'].min()

# انتخاب 200 نمونه تصادفی
df_subset = df.sample(n=200, random_state=42)

# مرتب‌سازی بر اساس زمان نسبی به صورت صعودی
df_sorted = df_subset.sort_values('RELATIVE_TIME', ascending=True)

# گروه‌بندی بر اساس شناسه ماشین (vehicleId)
grouped = df_sorted.groupby('vehicleId')

# ایجاد گراف شبکه
G = nx.Graph()

# حرکت روی حسگرها براساس ماشین‌ها
for group, data in grouped:
    # تعیین دوربین‌ها
    cameras = data['DEVICEID'].unique()

    # اضافه کردن گره‌ها به گراف
    for camera in cameras:
        G.add_node(camera)

    # اضافه کردن لینک‌ها بین دوربین‌ها با وزن برابر تعداد تکرارشان
    for i in range(len(cameras) - 1):
        camera1 = cameras[i]
        camera2 = cameras[i + 1]
        weight = len(data[(data['DEVICEID'] == camera1) & (data['DEVICEID'].shift(-1) == camera2)])
        G.add_edge(camera1, camera2, weight=weight)

# تنظیمات نمودار
pos = nx.spring_layout(G, seed=42)  # تعیین موقعیت گره‌ها با الگوریتم Spring
weights = [G[u][v]['weight'] for u, v in G.edges()]  # وزن لینک‌ها

# رسم گراف
nx.draw_networkx(G, pos, width=weights, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray')

# نمایش وزن لینک‌ها
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# نمایش نمودار
plt.title('Camera Sequence Network')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# خواندن دیتاست
df = pd.read_csv('Data_p1.csv')

# محاسبه زمان نسبی از صفر
df['RELATIVE_TIME'] = df['PASSDATETIME'] - df['PASSDATETIME'].min()

# انتخاب 200 نمونه تصادفی
df_subset = df.sample(n=200, random_state=42)

# مرتب‌سازی بر اساس زمان نسبی به صورت صعودی
df_sorted = df_subset.sort_values('RELATIVE_TIME', ascending=True)

# گروه‌بندی بر اساس دوربین
grouped = df_sorted.groupby('DEVICEID')

# تعداد تکرارهای هر دوربین
camera_counts = grouped.size()

# نمایش نمودار میله‌ای
plt.bar(camera_counts.index, camera_counts.values)
plt.xlabel('Camera ID')
plt.ylabel('Weight')
plt.title('Camera Sequence Weight')
plt.xticks(rotation=90)
plt.show()
```
